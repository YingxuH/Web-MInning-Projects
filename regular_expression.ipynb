```python
data = data.drop_duplicates(['review_id'])
data.columns.values.tolist()
```

```python
title_text = '\n'.join(data['product_title'].values.tolist())
```

```python
## basic sequence
earbud = data[data['product_title'].str.contains(r'earbud',regex=True)]
print(earbud.shape)
earbud.loc[earbud.index[-1],'product_title']
```

```python
# square braces for disjunction
earbud_ignorecase1 = data[data['product_title'].str.contains(r'[eE]arbud',regex=True)]
print(earbud_ignorecase1.shape)
earbud_ignorecase1.loc[earbud_ignorecase1.index[-1],'product_title']
```

```python
# case insensitive
earbud_ignorecase2 = data[data['product_title'].str.contains(r'(?i)[eE]arbud',regex=True)]
print(earbud_ignorecase2.shape)
earbud_ignorecase2.loc[earbud_ignorecase2.index[-1],'product_title']
```

```python
earbud_ignorecase2[~earbud_ignorecase2['review_id'].isin(earbud_ignorecase1['review_id'].values.tolist())]['product_title'].values.tolist()[0]
```

```python
bool(re.search(r'(?i)[eE]arbud', "Gear Beast GearBuds Noise Reducing Stereo In-Ear headphones"))
```

```python
# word boundary
earbud_ignorecase2 = data[data['product_title'].str.contains(r'\b(?i)[eE]arbud\b',regex=True)]
print(earbud_ignorecase2.shape)
earbud_ignorecase2.loc[earbud_ignorecase2.index[-1],'product_title']
```

```python
# quantifiers
earbuds = data[data['product_title'].str.contains(r'\b(?i)earbuds?\b',regex=True)]
print(earbuds.shape)
```

```python
# quantifiers, +, *, {}
# digits, \d, \w, \s(line break, tab, space)
model_digits = data[data['product_title'].str.contains(r'(?i)\b\w\d*\w\b',regex=True)]
print(model_digits.shape)
model_digits.loc[model_digits.index[-1],'product_title']
```

```python
# anchors
data['review_body'] = data['review_body'].fillna('')
review_the = data[data['review_body'].str.contains(r'(?i)^The',regex=True)]
print(review_the.shape)
review_the.loc[review_the.index[-1],'review_body']
```

```python
# anchors
data['review_body'] = data['review_body'].fillna('')
review_exclam = data[data['review_body'].str.contains(r'(?i)^The.*!$',regex=True)]
print(review_exclam.shape)
review_exclam.loc[review_exclam.index[-2],'review_body']
```

```python
# or, earbud or Earbud
earbud_ignorecase3 = data[data['product_title'].str.contains(r'e|Earbud',regex=True)]
print(earbud_ignorecase3.shape)
earbud_ignorecase3.loc[earbud_ignorecase3.index[-1],'product_title']
```

```python
# group
earbud_ignorecase3 = data[data['product_title'].str.contains(r'(e|E)arbud',regex=True)] #[eE]arbud # 
print(earbud_ignorecase3.shape)
earbud_ignorecase3.loc[earbud_ignorecase3.index[-1],'product_title']
```

```python
# group terms, earbud or earphone
earbud_phone = data[data['product_title'].str.contains(r'(?i)ear(bud|phone)',regex=True)]
print(earbud_phone.shape)
earbud_phone.loc[earbud_phone.index[-1],'product_title']
```

```python
# group terms, earbud or earphone
bla2 = data[data['review_body'].str.contains(r'(<br\s/>){2,}',regex=True)]
print(bla2.shape)
bla2['extract'] = bla2['review_body'].str.extract(r'(<br\s/>){2,}')
bla2.loc[bla2.index[-1],'review_body']
```

```python
# greedy versus lazy match, .* 
text = review_exclam.loc[review_exclam.index[-2],'review_body']
# print(text)
result = re.search(r'The.*!',text)
print("====greedy match====")
print(result.group())
result = re.search(r'The.*?!',text)
print("====lazy match====")
print(result.group())
```

```python
# multiline mode
text = "These are lovely!\nI just went over to [...] and found a cute template for package tags.These look so nice! I love these!"
print(text)
result = re.search(r'(?i)^The.*!$',text,flags=re.M)  # (?m) (?i)
print(result.group())
result = re.search(r'^The.*!$',text,flags=re.M)
print(result.group())
```

```python
# backreference
title = review_exclam.loc[review_exclam.index[-2],'product_title']
text = review_exclam.loc[review_exclam.index[-2],'review_body']
print("title: {}\n".format(title))
print(text+"\n")
result = re.search(r'These.*?!',text)
print("Match results:\n{}".format(result.group()))
result = re.search(r'(These).*?!',text)
print("Match results:\n{}".format(result.group()))
print("Match results group 0:\n{}".format(result.group(0)))
print("Match results group 1:\n{}".format(result.group(1)))
```

```python
# backreference in the same pattern
data['review_body'] = data['review_body'].fillna('')
quality = data[data['review_body'].str.contains(r'(?i)(good)(\w|\s)+\1',regex=True)]
quality.loc[quality.index[-1],'review_body']
extracts = quality['review_body'].str.extract(r'(?i)(good)((\w|\s)+)(\1)')
print(quality.loc[quality.index[0],'review_body'])
print(extracts.iloc[0,])
extracts = quality['review_body'].str.extract(r'(?i)(?P<first_good>good)(?P<Middle>(\w|\s)+)(?P<second_good>\1)')
# extracts.iloc[:3,]
quality = data[data['review_body'].str.contains(r'(?i)(?:works\s)(good)(\w|\s)+\1',regex=True)]
# quality['extract']
print(quality.shape)
extracts = quality['review_body'].str.extract(r'(?i)(?:works\s)(?P<first_good>good)(?P<Middle>(\w|\s)+)(?P<second_good>\1)')
extracts.iloc[:3,]
```

```python
# backreference and replace
title = review_exclam.loc[review_exclam.index[-2],'product_title']
text = review_exclam.loc[review_exclam.index[-2],'review_body']
matched_text = re.findall(r'(?i)(these).*?!', text)
print("original text:\n{}\n\n".format(text))
replaced_text = re.sub(r'(?i)(these).*?(!)', r"\2", text)
print("replaced text:\n{}\n\n".format(replaced_text))
```

```python
# other functions in re
text = review_exclam.loc[review_exclam.index[-2],'review_body']
text_search = re.search(r'lovely',text)
if text_search is not None:
    print("search.{}".format(text_search.group()))
text_match = re.match(r'lovely',text)
if text_match is not None:
    print("match.{}".format(text_match.group()))
text_find = re.findall(r'lovely',text)
if text_find is not None and text_find!=[]:
   print("findall.{}".format(text_find))
```
