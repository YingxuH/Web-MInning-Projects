```python
import pandas as pd
import numpy as np
import re
```

```python
# Assignment 1
# identify incentivized reviews
incentivized_example = pd.read_csv('incentivized_examples.csv')
data = pd.read_csv("data.csv")
```

```python
incentivized_example = incentivized_example.dropna()
data = data.dropna(subset=['review_body'])

# drop the rows with the same discount text and review text (should I?)
incentivized_example = incentivized_example[incentivized_example['review_body'] != incentivized_example['discount_text']]
```

```python
incentivized_example['discount_text'].tolist()
```

```python
particle_string = "(received?|provided?|got(ten)?)"
incentives = "(samples?|products?|items?|gifts?|units?)"
adverb = "(free|at no charge)"
prep = "(for|in exchange (for|of)|to)"
purpose = "review"
```

```python
identifying_string = "(?i){}.*({}|{}.*{})".format(particle_string, adverb, prep, purpose)
```

```python
incentivized_example['bool'] = incentivized_example['review_body'].str.contains(identifying_string)
```

```python
incentivized_example[incentivized_example['bool']]
```

```python
incentivized_data = data[data['review_body'].str.contains(identifying_string)]
```
